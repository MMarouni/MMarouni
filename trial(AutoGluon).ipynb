```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
```

```python
df=df.sample(frac=1)
df_train,df_test=train_test_split(df,test_size=0.3,random_state=1)
# df_train.shape
# df_test.shape
```

```python
df_test.head()
```

```python
test_Data=df_test.drop(['Output'],axis=1)
test_Data.head()
```

```python
predictor= TabularPredictor(label ="Output").fit(train_data = df_train, verbosity = 2,presets='best_quality')
```

```python
predictor.fit_summary()
```

```python
predictor.leaderboard(df_train, silent=True)
```

```python
predictor.feature_importance(data=df_train)
```

```python
y_pred=predictor.predict(test_Data)
```

```python
y_pred
```

```python
test_acc = predictor.evaluate(df_test)
# from sklearn.metrics import mean_absolute_percentage_error
# # mean_absolute_percentage_error()
```
